```python
from PIL import Image, ImageDraw
import random
import math

# Function to draw eyes
def draw_eye(draw, center, diameter, pupil_diameter, iris_diameter):
    # Draw outer circle (eye)
    draw.ellipse([center[0] - diameter // 2, center[1] - diameter // 2,
                  center[0] + diameter // 2, center[1] + diameter // 2], fill=(255, 255, 255))
    
    # Randomize pupil center within the eye
    pupil_center = (random.randint(center[0] - diameter // 4, center[0] + diameter // 4),
                    random.randint(center[1] - diameter // 4, center[1] + diameter // 4))
    
    # Ensure iris diameter is smaller than eye diameter and larger than pupil diameter
    iris_diameter = min(max(iris_diameter, pupil_diameter + 10), diameter - 10)  # Adding 10 as a margin for pupil and subtracting 10 as a margin for iris
    
    # Draw pupil
    draw.ellipse([pupil_center[0] - pupil_diameter // 2, pupil_center[1] - pupil_diameter // 2,
                  pupil_center[0] + pupil_diameter // 2, pupil_center[1] + pupil_diameter // 2], fill=(0, 0, 0))
    
    # Draw iris as a black-bordered circle
    iris_radius = iris_diameter // 2
    iris_center = pupil_center
    iris_border_radius = (pupil_diameter + iris_diameter) // 4
    iris_fill_color = (0, 0, 0)  # Black color for iris
    draw.ellipse([iris_center[0] - iris_radius, iris_center[1] - iris_radius,
                  iris_center[0] + iris_radius, iris_center[1] + iris_radius], outline=(0, 0, 0))
    draw.ellipse([iris_center[0] - iris_border_radius, iris_center[1] - iris_border_radius,
                  iris_center[0] + iris_border_radius, iris_center[1] + iris_border_radius], fill=iris_fill_color)
    
    # Return bounding box of the eye
    return (center[0] - diameter // 2, center[1] - diameter // 2, center[0] + diameter // 2, center[1] + diameter // 2)

import math

# Function to calculate intersection points between a line segment and a circular arc
import math

def intersect_line_circle(x1, y1, x2, y2, cx, cy, r):
    dx = x2 - x1
    dy = y2 - y1
    dr = math.sqrt(dx**2 + dy**2)
    D = x1*y2 - x2*y1
    discriminant = r**2 * dr**2 - D**2
    
    if discriminant < 0:
        # No intersection
        return None
    
    sign_dy = 1 if dy >= 0 else -1
    x_intersect1 = (D*dy + sign_dy*dx*math.sqrt(discriminant)) / dr**2
    x_intersect2 = (D*dy - sign_dy*dx*math.sqrt(discriminant)) / dr**2
    
    sign_dx = 1 if dx >= 0 else -1
    y_intersect1 = (-D*dx + abs(dy)*math.sqrt(discriminant)) / dr**2
    y_intersect2 = (-D*dx - abs(dy)*math.sqrt(discriminant)) / dr**2
    
    return ((x_intersect1, y_intersect1), (x_intersect2, y_intersect2))




# Function to draw mouths as filled moon crescents with random rotation and teeth
def draw_mouth(draw, center, width, height, curvature1, curvature2, rotation_angle, num_teeth):
    # Calculate the coordinates of the four endpoints of the mouth arcs
    x1 = center[0] - width // 2
    x2 = center[0] + width // 2
    y1 = center[1] - height // 2
    y2 = center[1] + height // 2
    
    # Calculate the control points for the bezier curves
    cx1 = x1 + curvature1 * width
    cx2 = x2 - curvature1 * width
    cy1 = center[1] - height // 4
    cy2 = center[1] - height // 4
    
    # Draw the first arc with rotation
    draw.pieslice([x1, y1, x2, y2], rotation_angle, rotation_angle + 180, fill=(255, 255, 255))
    
    # Draw the second arc with rotation
    draw.pieslice([x1, y1, x2, y2], rotation_angle, rotation_angle - 180, fill=(255, 255, 255))
    
    # Calculate the angle between each tooth
    tooth_angle = 180 / (num_teeth + 1)
    
    # Calculate the angle for each tooth
    tooth_angles = [rotation_angle + (i + 1) * tooth_angle for i in range(num_teeth)]
    
    # Calculate the coordinates of the endpoints of the teeth segments
    for angle in tooth_angles:
        x1_tooth = center[0] + (width // 2) * math.cos(math.radians(angle))
        y1_tooth = center[1] + (height // 2) * math.sin(math.radians(angle))
        x2_tooth = center[0] + ((width // 2) + 10) * math.cos(math.radians(angle))
        y2_tooth = center[1] + ((height // 2) + 10) * math.sin(math.radians(angle))
        
        # Calculate intersection points with the first arc
        intersection1 = intersect_line_circle(x1_tooth, y1_tooth, x2_tooth, y2_tooth, x1, center[1], width // 2)
        
        # If intersection points exist, draw the tooth segment
        if intersection1 is not None:
            (x1_intersect, y1_intersect), (x2_intersect, y2_intersect) = intersection1
            draw.line([x1_intersect, y1_intersect, x2_intersect, y2_intersect], fill=(0, 0, 0), width=2)
        
        # Calculate intersection points with the second arc
        intersection2 = intersect_line_circle(x1_tooth, y1_tooth, x2_tooth, y2_tooth, x2, center[1], width // 2)
        
        # If intersection points exist, draw the tooth segment
        if intersection2 is not None:
            (x1_intersect, y1_intersect), (x2_intersect, y2_intersect) = intersection2
            draw.line([x1_intersect, y1_intersect, x2_intersect, y2_intersect], fill=(0, 0, 0), width=2)

# Create a new blank image
image = Image.new("RGB", (1000, 1000), "black")
draw = ImageDraw.Draw(image)

# Lists to store bounding boxes of eyes and mouths
eye_boxes = []
mouth_boxes = []

# Draw eyes
for _ in range(10):  # Draw 5 eyes
    attempts = 0
    while attempts < 10:  # Limit the number of attempts
        center = (random.randint(50, 950), random.randint(50, 950))
        diameter = random.randint(50, 100)
        pupil_diameter = random.randint(10, 30)
        iris_diameter = random.randint(pupil_diameter + 10, diameter - 10)  # Ensure iris diameter is within the specified range
        eye_box = draw_eye(draw, center, diameter, pupil_diameter, iris_diameter)
        overlap = False
        for box in eye_boxes + mouth_boxes:
            if (eye_box[0] < box[2] and eye_box[2] > box[0] and
                eye_box[1] < box[3] and eye_box[3] > box[1]):
                overlap = True
                break
        if not overlap:
            eye_boxes.append(eye_box)
            break
        attempts += 1

# Draw mouths with teeth
for _ in range(10):  # Draw 5 mouths
    attempts = 0
    while attempts < 30:  # Limit the number of attempts
        center = (random.randint(100, 900), random.randint(300, 900))
        width = random.randint(30, 300)
        height = random.randint(10, 400)
        curvature1 = random.uniform(0.1, 0.5)
        curvature2 = random.uniform(0.1, 0.5)
        num_teeth = random.randint(3, 14)  # Random number of teeth
        mouth_box = (center[0] - width // 2, center[1] - height // 2, center[0] + width // 2, center[1] + height // 2)
        rotation_angle = random.uniform(0, 360)
        overlap = False
        for box in eye_boxes + mouth_boxes:
            if (mouth_box[0] < box[2] and mouth_box[2] > box[0] and
                mouth_box[1] < box[3] and mouth_box[3] > box[1]):
                overlap = True
                break
        if not overlap:
            draw_mouth(draw, center, width, height, curvature1, curvature2, rotation_angle, num_teeth)
            mouth_boxes.append(mouth_box)
            break
        attempts += 1

# Display the image
image
```
